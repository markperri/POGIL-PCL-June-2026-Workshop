### Pandas is the python library for working with data. It's basically the python equivalent of excel.
Instead of Excel worksheets we have dataframes (df)

CSV files can be imported into a dataframe, just as you import a CSV into Excel


```python
import pandas as pd
df = pd.read_csv("cyanine.CSV")

# if the last line in a cell is a variable name, it will display it for you
df
```

### To plot data from a dataframe we use matplotlib, a python plotting library
We can reference a column in an dataframe using df['column name']

This is similar to Excel where you can reference a different worksheet:  =Sheet2!A:A

```python
# plot absorbance vs wavelength
import matplotlib.pyplot as plt

wavelength = df['Wavelength (nm)']
absorbance = df['Absorbance (AU)']

plt.figure(figsize=(8, 5))
plt.plot(wavelength, absorbance)
plt.xlabel('Wavelength / nm')
plt.ylabel('Absorbance / AU')
plt.title('UV-Vis of cyanine dye')
plt.legend()
plt.grid(True)
plt.show()
```

### Now let's have python find the peaks for us

```python
from scipy.signal import find_peaks
indices, properties = find_peaks(absorbance, prominence=0.02)

indices
```

### Subset our data corresponding to these peaks

```python
# take a subset of our data corresponding to these peaks
# we are looking for rows this time, so we use .iloc (index location)
peaks_df = df.iloc[indices]

peaks_df
```

```python
# plot the peaks overlayed on our original spectrum
peak_wavelengths = peaks_df['Wavelength (nm)']
peak_absorbances = peaks_df['Absorbance (AU)']

plt.figure(figsize=(8, 5))
plt.plot(wavelength, absorbance)
plt.plot(peak_wavelengths, peak_absorbances, 'ro')

plt.xlabel('Wavelength / nm')
plt.ylabel('Absorbance / AU')
plt.title('UV-Vis of cyanine dye')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Smooth the data using a moving average

from scipy.signal import savgol_filter  #Savitzky-Golay filter

# window_length must be odd, otherwise peak will shift to one side.
# polyorder: the polynomial degree to fit the data to

# important:  the moving average will alter the height, so be careful if that's important to you!
smoothed_absorbance = savgol_filter(absorbance, window_length=15, polyorder=2)

plt.figure(figsize=(8, 5))
plt.plot(wavelength, absorbance, 'b-', alpha=0.3, label='Raw Data')
plt.plot(wavelength, smoothed_absorbance, 'g-', label='Smoothed Data')

plt.xlabel('Wavelength / nm')
plt.ylabel('Absorbance / AU')
plt.title('Smoothed data')
plt.legend()
plt.grid(True)
plt.show()

```

```python
# find the peaks again, using this smoothed dataset
indices, properties = find_peaks(smoothed_absorbance, prominence=0.02)
peaks_df = df.iloc[indices]

peaks_df
```

```python
# plot everything!

plt.figure(figsize=(8, 5))
plt.plot(wavelength, absorbance, 'b-', alpha=0.3, label='Raw Data')
plt.plot(wavelength, smoothed_absorbance, 'g-', label='Smoothed Data')
plt.plot(peaks_df['Wavelength (nm)'], smoothed_absorbance[indices], 'ro', label='Peaks')

plt.xlabel('Wavelength / nm')
plt.ylabel('Absorbance / AU')
plt.title('Peak Detection with Savitzky-Golay Smoothing')
plt.legend()
plt.grid(True)
plt.show()
```
