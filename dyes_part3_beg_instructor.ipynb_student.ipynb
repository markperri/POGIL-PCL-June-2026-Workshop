
$$\text{Experimentally we found the wavelengths of these transitions, which we can convert to energy:}$$
$$\Delta E = \frac{hc}{\lambda}$$
$$\text{where:}$$


$$\Delta E = \left(E_{\text{LUMO}} - E_{\text{HOMO}}\right)$$
$$\text{and that energy can be substituted into the particle in a box equation to calculate the effective length of the box:}$$


$$E = \frac{h^2}{8mL^2} (n^2)$$

$$\text{So:}$$

$$\Delta E = \frac{h^2}{8mL^2} \left(n_{\text{LUMO}}^2 - n_{\text{HOMO}}^2\right)$$

$$L^2 = \frac{h^2}{8m\Delta E} \left(n_{\text{LUMO}}^2 - n_{\text{HOMO}}^2\right)$$

$$L = \sqrt{\frac{h^2}{8m\Delta E} \left(n_{\text{LUMO}}^2 - n_{\text{HOMO}}^2\right)}$$




<center>
<p>The "box" is the central part of the molecule between the two nitrogens.  It does not count the side groups.</p>

<p>We can get the index of the HOMO by counting electrons: 1 for each carbon between the nitrogens and 3 total for the two nitrogens</p>

```python
# calculate the index of the HOMO and LUMO
n_HOMO = electrons / 2
n_LUMO = n_HOMO + 1

# the number of carbon atoms = # electrons - 3 (each C contributes 1 electron and the two nitrogens contribute 3)
carbons = electrons - 3

# the number of bonds = # carbon atoms + 1
bonds = carbons + 1

L = np.sqrt((h**2 / (8 * m * delta_E)) * (n_LUMO**2 - n_HOMO**2))


# optional:  put it in a dataframe to display it nicely
import pandas as pd
df = pd.DataFrame({
	'wavelength (nm)': wavelength_nm,
	'∆E (J)': delta_E,
	'box length (m)': L,
	'carbon atoms': carbons,
	'bonds': bonds, 
	'electrons': electrons, 

	})
pd.set_option('display.precision', 2)
display(df)
```

The following cells calculate the fit from Moog, *J. Chem. Educ.*, 1991. The two parameter fit is for the average bondlength in the conjugated chain and a parameter, $\gamma$, that represents the additional length due to the polarizability at the ends of the chain.

$$L = (b \times l) + \gamma$$

where $b$ and $L$ are the number of bonds calculated box length for each dye, $l$ is the average bondlength, and $\gamma$ is the additional length.
